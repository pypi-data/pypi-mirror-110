```python


import mobster as mb

import numpy as np
```

```python
data = mb.load_example_data("./")

#data = data["1:1"]
```

```python
fitted = mb.fit_mobster(data,1,1,purity = 1,  truncated_pareto=True, max_it = 200, lr = 0.05,  e = 0.0001, lrd_gamma = 0.1,   
        number_of_trials_k= 100, prior_lims_k=[0.1,10000], prior_lims_clonal=[0.1,10000.], number_of_trials_clonal_mean= 300.)
```

```python

mb.plot_results(data, fitted, output= "cane.png", drivers=None)
```
