```python
import os
from torch.utils.data import DataLoader
from earthvision.datasets import AerialCactus
```

```python
train_dataset = AerialCactus(root=os.getcwd(), data_mode='training_set')
valid_dataset = AerialCactus(root=os.getcwd(), data_mode='validation_set')
```

```python
train_dataloader = DataLoader(train_dataset, batch_size=64, shuffle=True)
valid_dataloader = DataLoader(valid_dataset, batch_size=64, shuffle=True)
```

```python
train_data, train_label = next(iter(train_dataloader))
valid_data, valid_label = next(iter(valid_dataloader))
```

```python
print(train_data.shape)
print(train_label.shape)
print(valid_data.shape)
print(valid_label.shape)
```
